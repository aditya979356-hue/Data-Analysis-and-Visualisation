## Data Analysis and Data Visualization

## importing Libraries

```python
import pandas as pd
import csv
import matplotlib.pyplot as plt
import seaborn as sns
```

## Load a CSV File


```python


try:
    file = open(r"EmpList.csv","r",newline="")
    df=pd.read_csv(file)
    print("CSV File Loaded Successfully")
except FileNotFoundError:
    print("Error : 'Your file is not found'")
    Data = {
        "Name":["Aditya","Ajay","John","Ibrahim"],
        "Employee_ID":[112,113,114,115]
    }
    df = pd.DataFrame(Data)
```

## Basic Data Analysis with Pandas

```python
print("\n ---Basic DATA Analysis---")

print("\n DataFrame Information:")
print(df.info())

print("\n Descriptive Statistics:")
print(df.describe())

print("\n Display Fist 10 rows of fie")
print(df.head(10))

print("\n Display Last 10 rows of File")
print(df.tail(10))
```

## Filtering Rows and Adding/Removing and Updating Columns 

```python
filteredRows = df[(df["SALARY"]<10000) & (df["SALARY"]>5000)]
print(filteredRows)
# print(df["FIRST_NAME"])

#Adding Columns by two Methods
#df.insert(12,"Bonus",df["SALARY"]*0.1)
# df["Bonus"]=df["SALARY"]*0.05

#Updating Columns
for i in range(0,49):
    df["Bonus"] = df["SALARY"]*0.001*i

#Removing Columns
# df.drop(columns=["Bonus"], inplace= False)
df
```

## HANDLING MISSING DATA

```python
df.isnull()
df.interpolate(method="linear",axis=0,inplace=False)
```

## Basic Aggregations :- sum(),mean(),min(),max(),count(),std()

```python
print("Total Salary given to Employees:",df["SALARY"].sum())
print("Highest Salary given to Employees:",df["SALARY"].max())
print("Lowest Salary given to Employees:",df["SALARY"].min())
print("Average Salary given to Employees:",df["SALARY"].mean())
print("Standard Deviation in Salary given to Employees:",df["SALARY"].std())
print("Lowest Salary given to Employees:",df[df["SALARY"]>10000].count())
```

## GROUP BY


```python
#Average Salary given to male and female
df.groupby(["Gender"])["SALARY"].mean()
```

## DATA VISUALISATION

```python
# Bar Chart
plt.title("AGE VS SALARY COMPARISION")
x = df["Age"]
y = df["SALARY"]
plt.bar(x,y,color="orange",label="Salary")
plt.xlabel("AGE")
plt.ylabel("SALARY")
age = [15,20,25,30,35,40,45,50,55,60,65,70]
plt.xticks(age)
plt.legend()
plt.savefig("barChart.png",dpi = 300,bbox_inches="tight")
plt.show()
```

```python
#Scatter Plot
x = df["Age"]
y = df["SALARY"]
plt.scatter(x ,y ,marker="o" , label = "SALARY")
plt.title("Age vs Salary Comparision")
plt.xlabel("Age")
plt.ylabel("Salary")
age = [15,20,25,30,35,40,45,50,55,60,65,70]
plt.xticks(age)
plt.legend()
plt.savefig("scatterChart.png",dpi = 300,bbox_inches="tight")
plt.show()
```

```python
#MULTIPLE CHARTS IN SINGLE CHARTS

plt.subplot(1,2,1)
# Bar Chart
plt.title("AGE VS SALARY COMPARISION Bar")
x = df["Age"]
y = df["SALARY"]
plt.bar(x,y,color="orange",label="Salary")
plt.xlabel("AGE")
plt.ylabel("SALARY")
age = [15,20,25,30,35,40,45,50,55,60,65,70]
plt.xticks(age)
plt.legend()


plt.subplot(1,2,2)
#Scatter Plot
x = df["Age"]
y = df["SALARY"]
plt.scatter(x ,y ,marker="o" , label = "SALARY")
plt.title("Age vs Salary Comparision Scatter")
plt.xlabel("Age")
plt.ylabel("Salary")
age = [15,20,25,30,35,40,45,50,55,60,65,70]
plt.xticks(age)
plt.legend()
plt.tight_layout()
plt.savefig("barAndscatterChart.png",dpi = 300,bbox_inches="tight")
plt.show()
```

```python
#HeatMAP
columns = [ "Age","SALARY"]

sns.heatmap(data=df[columns].corr(),annot=True)
plt.title("AGE VS SALARY COMPARISION BY HEATMAP")
plt.savefig("heatMap.png",dpi = 300,bbox_inches="tight")
```

```python
#Scatter Plot using seaborn
plt.title("AGE VS SALARY COMPARISION BY GENDER")
sns.scatterplot(data=df,x="Age",y = "SALARY",hue = "Gender")
plt.savefig("scatterPlot.png",dpi = 300,bbox_inches="tight")
```
